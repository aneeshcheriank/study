```python
import sys
import os
import glob
import openai


from langchain.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.embeddings.openai import OpenAIEmbeddings
from langchain.vectorstores import Chroma

if '../../auth' not in sys.path:
    sys.path.append('../../auth')
    
import openai_auth

DOC_FOLDER = './doc'
openai.api_key = os.environ['OPENAI_API_KEY']
CHUNK_SIZE = 1000
CHUNK_OVERLAP = 150
VECTOR_DIR = './out/vector_dir'


def get_files(doc_folder):
    return glob.glob(f'{doc_folder}/*')

def split_docs(file_path):
    
    # load pdf documents
    loader = PyPDFLoader(file_path)
    documents = loader.load()
    
    # split the documents to chunks
    text_splitter = RecursiveCharacterTextSplitter(chunk_size = CHUNK_SIZE, chunk_overlap = CHUNK_OVERLAP)
    docs = text_splitter.split_documents(documents)
    
    return docs

def add_docs_to_db(docs, vectordb=vectordb, persist_dir=VECTOR_DIR):
    # adding documents to the vectordb
    vectordb.add_documents(docs)
    
    print(f'vector store docs count: {vectordb._collection.count()}')
```

```python
if __name__ == '__main__':
    
    embedding = OpenAIEmbeddings()
    vectordb = Chroma.from_documents(
        documents = docs,
        embedding = embedding,
        persist_directory = persist_dir
    )
    
    files = get_files(DOC_FOLDER)
    docs = split_docs(files[0])
    add_docs_to_db(docs)

    len(docs)
```
