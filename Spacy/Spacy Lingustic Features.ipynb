### POS tagging
- word fall into which of the Part of speach
- noun, pronoun, verb, adverb, adjective, conjunction, preposition, interjunction, determiner

    - Determiner: Is placed in front of a noun to express a quantity or clarify what the noun refers tobriefly, a noun introducer
    - Preposition: Connects a noun/pronoun to other parts of the sentence
    - Conjunction: Glues words, clauses, and sentences together
    - Interjection: Expresses emotion in a sudden and exclamatory way
    
### universal tags
- pos_ feature


- `spacy.explain()`
    - explain the POS tag
    - 2 ways to use
        - use on tags
        - use on tag names
        - example given below
    - <b>pos_, tag_</b>

```python
import spacy

nlp = spacy.load('en_core_web_md')

print(spacy.explain("NNS"))

doc = nlp('I saw flowers.')
token = doc[2]
token.text, token.tag_, spacy.explain(token.tag_)
```

```python
import spacy

nlp = spacy.load('en_core_web_md')

doc = nlp('My cat will fish for a fish tomorrow in a fishy way.')
for token in doc:
    print(
        (token.text, token.pos_, token.tag_, 
        spacy.explain(token.pos_), spacy.explain(token.tag_))
    )
```

## Dependency parsing
- sentence structure via dependencies between tokens
- a tree

### dependency relationships (Labels)


```python
import spacy
from spacy import displacy

nlp = spacy.load('en_core_web_md')

doc = nlp('My cat will fish for a fish.')
displacy.render(doc, style='dep')
```
