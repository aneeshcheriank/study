                - news: news, media

## Visualization with Spacy

- displaCy
- Entity visualizer

```python
import spacy
from spacy import displacy

nlp = spacy.load('en_core_web_md')

doc = nlp('I have a ginger cat.')
displacy.serve(doc, style='dep')
# can add a port parameter in the serve
# displacy.serve(doc, style='dep', port='5001')
```

```python
import spacy
from spacy import displacy

nlp = spacy.load('en_core_web_md')

txt = 'Satya Nadalla is the CEO of Microsoft.'

doc = nlp(txt)
displacy.serve(doc, style='ent')
```

```python
import spacy
from spacy import displacy

nlp = spacy.load('en_core_web_md')

txt = 'Satya Nadalla is the CEO of Microsoft.'

doc = nlp(txt)
displacy.render(doc, style='ent')
# .reder will work on the jupyer-notebook
```
