## Customizing tokenizer

- some domains like finance, insurace have some special words
- need to customize the tokenizer
- can add special tokenization rules
- example code
- <b>Note</b>: add the special rules only if necessary. It may lead to undesired results

```python
import spacy
from spacy.symbols import ORTH

nlp = spacy.load('en_core_web_md')

txt = 'lemme that'

doc = nlp(txt)
print([w.text for w in doc])

# adding special case to the tokenizer
special_case = [{ORTH: 'lem'}, {ORTH: 'me'}]
nlp.tokenizer.add_special_case('lemme', special_case)
print([w.text for w in nlp(txt)])
```
